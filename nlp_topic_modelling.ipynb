The function below anonymises data by removing personal names and all the gym chain & gym words (including their spelling variants) from text in the reviews.

The reasons behind this approach are:

* Keep Privacy: hiding people's names (e.g., "Ellie") by replacing them with "[NAME]"
* Refine Focus: Removed gym-specific words (like "the gym chain," "Gym," or "fitness club") so the topic modelling focuses on emotions and general themes, rather than specific places. Otherwise, the word "Gym" would be repeated unnecessarily throughout the analysis, adding no value to the models.
* This approach helps to identify feelings (e.g. anger) in the reviews without being thrown off by names or gym details.

```python
# Anonymisation of the data sets to remove any personal information


# Disclosure: The following function was created with AI's assistance
nlp = spacy.load("en_core_web_sm")

def anonymize_names_simple(df, text_columns):
    """
    Anonymize names in specified DataFrame columns using NER and a fallback name list,
    remove gym-related terms, and clean up extra spaces.

    Parameters:
    - df: pandas DataFrame containing the text data
    - text_columns: String or list of column names to anonymize

    Returns:
    - DataFrame with anonymized names and gym-related terms removed
    """
    # Ensure text_columns is a list
    if isinstance(text_columns, str):
        text_columns = [text_columns]
    if not all(col in df.columns for col in text_columns):
        raise ValueError(f"One or more columns {text_columns} not found in DataFrame")

    df_anon = df.copy()

    # Fallback list of common names (case-insensitive)
    common_names = {
        'ellie', 'kennedy', 'john', 'jane', 'mary', 'mari', 'smith', 'jones',
        'sarah', 'david', 'emma', 'michael', 'olivia', 'james', 'sophia', 'robert',
        'isabella', 'william', 'mia', 'daniel', 'ava', 'matthew', 'amelia', 'joseph'
    }
    name_pattern = r'\b(?:' + '|'.join(re.escape(name) for name in common_names) + r')\b'

    def replace_names_and_remove_gym(text):
        if not isinstance(text, str):
            return text

        # Step 1: Replace names using spaCy NER
        doc = nlp(text)
        for ent in doc.ents:
            if ent.label_ == "PERSON":
                text = text.replace(ent.text, '[NAME]')

        # Step 2: Fallback - Replace known names from the list (case-insensitive)
        text = re.sub(name_pattern, '[NAME]', text, flags=re.IGNORECASE)

        # Step 3: Remove gym-related terms
        gym_patterns = [
            r'\bthe gym chain\b', r'\bthe gym chain\b', r'\b[Chain]\b', r'\bthe gym chain\b',
            r'\bthe gym chain\b', r'\bPure\s+Gym\b', r'\bPURE\s*GYM\b',
            r'\bPure\s*Gym\b', r'\bgym\w*\b',
            r'\bfitness\s*center\b', r'\bfitness\s*centre\b',
            r'\bhealth\s*club\b', r'\bfitness\s*club\b',
            r'\bworkout\s*center\b', r'\bworkout\s*centre\b'
        ]
        combined_gym_pattern = '|'.join(gym_patterns)
        text = re.sub(combined_gym_pattern, '', text, flags=re.IGNORECASE)

        # Step 4: Clean up extra spaces
        text = re.sub(r'\s+', ' ', text).strip()

        return text

    # Apply the function to specified columns
    for col in text_columns:
        df_anon[col] = df_anon[col].apply(replace_names_and_remove_gym)
    return df_anon

# Implementation
google_df_anon = anonymize_names_simple(google_df, 'Comment')
trustpilot_df_anon = anonymize_names_simple(trustpilot_df, ['Review Title', 'Review Content'])

print("Anonymised Google DataFrame:")
print(google_df_anon.head(20))
print("\nAnonymised Trustpilot DataFrame:")
print(trustpilot_df_anon.head(20))
```

```python
# Criterion 2.9: Plot a histogram/bar plot showing the top 10 words from each data set
def plot_top_words(counter, title):
    words, counts = zip(*counter.most_common(10))
    plt.figure(figsize=(10, 6), facecolor=WHITE)
    sns.barplot(x=list(counts), y=list(words), color=NAVY_BLUE)
    plt.title(title, color=NAVY_BLUE, fontsize=14)
    plt.xlabel('Frequency', color=NAVY_BLUE)
    plt.ylabel('Words', color=NAVY_BLUE)
    plt.gca().set_facecolor(WHITE)
    plt.gca().tick_params(axis='x', colors=NAVY_BLUE)
    plt.gca().tick_params(axis='y', colors=NAVY_BLUE)
    plt.show()

# The plots highlight dominant themes for business insights
plot_top_words(google_word_freq, 'Top 10 Words in Google Reviews')
plot_top_words(trustpilot_word_freq, 'Top 10 Words in Trustpilot Reviews')
```

```python
### Conducting initial topic modelling

# Criterion 3.1: Filter out the reviews that are from the locations common to both data sets (negative reviews dataframes) and merge into a new list
common_neg_google = google_negative[google_negative["Club's Name"].isin(common_locations)] # filters the google negative reviews where the "Club's Name" is in common_locations
common_neg_trustpilot = trustpilot_negative[trustpilot_negative['Location Name'].isin(common_locations)] # as above but for Trustpilot + 'Location Name'
combined_common_neg = list(common_neg_google['Comment']) + list(common_neg_trustpilot['Review Content']) # After filtering, the comments from the two data sets are combined into a single list
```

```python
# Criterion 3.2: Preprocess this data set and use BERTopic on this cleaned data set
combined_common_neg = [preprocess_text(text) for text in combined_common_neg] # Each review in the combined_common_neg list is cleaned using preprocess_text(text)
combined_common_neg_subset = combined_common_neg[:1000]  # To speed up computation, only the first 1000 reviews are selected
embedding_model = SentenceTransformer("all-MiniLM-L6-v2", device='cpu') # all-MiniLM-L6-v2 is used to generate numerical embeddings for each review
embeddings = embedding_model.encode(combined_common_neg_subset, show_progress_bar=True) # converts the text into high-dimensional vectors
# Note: all-MiniLM-L6-v2 balances efficiency and quality for embedding generation
```

**BERTopic Model using KMeans**

In the code below, the model is using KMeans clustering over Dendrogram (as explained by the professor during the lessons); in this case KMeans is more appropiate because:

* Picked 10 topics (n_clusters=10) for easy results. Dendrograms, needed manual splitting, which is not clear
* It was quick and good for the data being analysed. Dendrogram was slow and used more resources
* Allowed the model to calculate probabilities for topic assignment

For this project, KMeans was faster, simpler, and better for getting 10 clear topics from negative reviews in BERTopic.

```python
# Creating a BERTopic Model
# BERTopic leverages contextual embeddings, outperforming other methods for nuanced topic extraction from negative reviews
topic_model = BERTopic(
    umap_model=umap.UMAP(n_neighbors=30, n_components=2, min_dist=0.0, metric='cosine'), # Configures UMAP for dimensionality reduction with specified parameters
    hdbscan_model=KMeans(n_clusters=10), # KMeans clustering to group data into 10 topics
    embedding_model=None,
    language="english", # sets the language of the text data
    calculate_probabilities=True, # calculates probabilities for topic assignment
    verbose=True
)

# Hyperparameters selection
    # n_neighbors=30, n_components=2 in UMAP reduce dimensionality effectively for clustering.
    # In this case, using n_neighbors=30 enabled the UMAP model to maintain a more generalised understanding of the data set's structure
    # n_clusters=10 as per requiremnt

# Model training
topics, probs = topic_model.fit_transform(combined_common_neg_subset, embeddings=embeddings) # topics lists where each element is a topic label and probs lists each element in a probability indicating the strength of the association with each topic
```

```python
# Criterion 3.3: List out the top topics along with their document frequencies
topic_info = topic_model.get_topic_info() # returns a pandas dataframe that contains metadata about all the topics identified by the BERTopic model
print("Top Topics with Document Frequencies:")
print(topic_info.head(10)) # displays the first 10 rows of the dataframe, which represent the top 10 topics
```

```python
# Criterion 6.1: Load Falcon-7B-Instruct
torch.cuda.empty_cache() # clears the GPU memory cache in PyTorch
falcon = pipeline("text-generation", model="tiiuae/falcon-7b-instruct", device=-1) # creates a text generation pipeline
print("Model loaded successfully")
```

```python
# Criteria 6.2–6.4: Process reviews
subset_anger_reviews = anger_reviews[:50] # Takes the first 50 reviews from anger_reviews
prompt = ( # template prompt that will be used with each review,  it instructs the model to analyse a customer review and identify exactly 3 main topics
    "Analyse the following customer review and identify exactly 3 main topics. "
    "Return them strictly in this exact format, with no additional text or deviation:\n"
    "1. [Topic]\n2. [Topic]\n3. [Topic]\n\n"
    "Review:\n"
)

# Please note that I had to tweak the prompt as I was getting very poor results
# with the original prompt, which was: "In the following customer review, pick out the main 3 topics. Return them in a numbered list format, with each one on a new line."

falcon_outputs = [] # Disclosure: parts of this code were created with AI's assistance
for i, review in enumerate(subset_anger_reviews): # iterates through each review in the subset, for each review, it combines the template prompt with the specific review text
    output = falcon( # calls the Falcon model with defined parameters
        prompt + review,
        max_new_tokens=500, # Limits the response to 500 new tokens (~ 375-500 words)
        temperature=0.7, # Sets moderate randomness in the generation (higher values = more creative/random)
        top_p=0.9, # Uses nucleus sampling, considering only the top 90% most probable next tokens
        do_sample=True # Enables probabilistic sampling rather than just taking the most likely token
    )[0]['generated_text']
    falcon_outputs.append(output)

    # Hyperparameters selection - after multiple trial and errors, and with AI's help, the hyperparameters were set as follows:
    # max_new_tokens=500: Limits output length for efficiency while ensuring detail
    # temperature=0.7, top_p=0.9: Balances creativity and coherence in topic/insight generation
```

```python
# Criterion 6.5: Topic extraction - Disclosure: parts of this code were created with AI's assistance
all_topics = [] # empty list to store all extracted topics
for i, output in enumerate(falcon_outputs): # Iterates through each model output with its index
    print(f"Raw Output {i+1}:", output) # Prints the raw output for reference
    lines = output.split('\n') # Splits the output into lines
    review_idx = next((idx for idx, line in enumerate(lines) if "Review:" in line), -1) # finds where the review text begins by locating the line containing "Review:"
    content_lines = lines[review_idx + 1:] if review_idx != -1 else lines #  isolates just the content after the "Review:" marker
    topics = re.findall(r'^\d+\.\s*\[?(.+?)\]?\s*$', '\n'.join(content_lines), re.MULTILINE) # find numbered lines
    topics = [t for t in topics if t.lower() != "topic"]
    if not topics:
        review = subset_anger_reviews[i]
        keywords = re.findall(r'\b(?:gym|staff|equipment|service|crowd|hygiene|clean|rude|machine)\w*\b', review, re.IGNORECASE)
        topics = list(dict.fromkeys(keywords))[:3]
    if len(topics) < 3:
        topics.extend([f"Concern {j+1} from review {i+1}" for j in range(3 - len(topics))])
    print(f"Extracted Topics {i+1}: {topics[:3]}")
    all_topics.extend(topics[:3])
print("All Topics:", all_topics)
print("Sample Extracted Topics:", all_topics[:10])
if not all_topics:
    raise ValueError("No topics extracted - extraction logic failed")

# This code extracts topics from falcon_outputs into all_topics, looping through
# each output to print raw text, split it into lines, and isolate content after
# "Review:". It uses regex to find numbered topics, or extracts up to three
# keywords (e.g., "gym," "staff") if none are found, adding placeholders
# if needed. Topics are printed, added to all_topics, and the code prints all
# topics plus a sample of the first 10, raising an error if no topics are extracted
```

```python
# Criteria 6.8–6.11: Generate insights - # Disclosure: parts of this code were created with AI's assistance
insight_prompt = (
    "Given the following topic cluster from negative customer reviews, provide exactly 3 actionable insights "
    "for the gym company. Return them strictly in this format, with no additional text:\n"
    "1. [Topic] - [Suggestion]\n"
    "2. [Topic] - [Suggestion]\n"
    "3. [Topic] - [Suggestion]\n\n"
    "Cluster:\n"
)

# Please note that here I also had to modify the prompt as I was getting unsuccessful results
# Original prompt: "For the following text topics obtained from negative customer reviews, can you give some actionable insights that would help this gym company?"

insights = []
cluster_info = topics_model.get_topic_info()
for _, row in cluster_info.iterrows():
    cluster_name = row['Name']
    output = falcon(
        insight_prompt + cluster_name,
        max_new_tokens=500,
        temperature=0.7,
        top_p=0.9,
        do_sample=True
    )[0]['generated_text']
    print(f"Cluster '{cluster_name}' Output:", output)
    # Relaxed parsing: match numbered lines, assume content is suggestion
    lines = output.split('\n')
    review_idx = next((idx for idx, line in enumerate(lines) if "Cluster:" in line), -1)
    content_lines = lines[review_idx + 1:] if review_idx != -1 else lines
    numbered_lines = [line.strip() for line in content_lines if line.strip().startswith(('1.', '2.', '3.'))]
    # Fallback: if no numbered lines, use raw lines and pair with cluster topics
    if not numbered_lines:
        cluster_topics = cluster_name.split('_')[1:]  # Extract topics from cluster name
        numbered_lines = [f"{i+1}. [{cluster_topics[0] if i < len(cluster_topics) else 'General Issue'}] - {line.strip()}"
                         for i, line in enumerate(content_lines[:3]) if line.strip()]
    insights.extend(numbered_lines[:3])
```
